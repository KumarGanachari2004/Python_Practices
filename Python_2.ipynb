```python
#Trailing Zeros
def trailing_zeros(n):
    count=0
    while n>0:
        n=n//5
        count+=n
    return count
n=int(input("Enter noumber:"))
print(f"tariling Numbers are:{trailing_zeros(n)}")
```

```python
#Iterative Power
def power(x,y):
    res=1
    for i in range(1,y+1):
        res=res*x
    return res
x=int(input("Enter noumber:"))
y=int(input("Enter noumber:"))
print(f"Power Numbers is:{power(x,y)}")
```

```python
#sum Of Digits
def sum(n):
    s=0
    while n!=0:
        d=n%10
        s=s+d
        n=n//10
    return s
n=int(input("Enter noumber:"))
print(f"Sum is:{sum(n)}")

        
```

```python
#Sum of even digits
def sum(n):
    s=0
    while n!=0:
        d=n%10
        if d%2==0:
            s=s+d
            n=n//10
    return s
n=int(input("Enter noumber:"))
print(f"Sum is:{sum(n)}")
```
