```python
#Max of two numbers
def max(a,b):
    if a>b:
        return a
    else:
        return b
a=int(input("Enter u r number"))
b=int(input("Enter u r number"))
print("Max Number is :",max(a,b))
```

```python
#Swap of two nors
def swap(a,b):
    a,b=b,a
    return a,b
a=int(input("Enter u r number"))
b=int(input("Enter u r number"))
print("Swap Number is :",swap(a,b))
```

```python
#absolute value
def abs(n):
    if n<0:
        return -n
    else:
        return n
n=int(input("Enter u r Number:"))
print(f"original value={n} & Abs Value={abs(n)}")
```

```python
#Sum of n Natural Numbers
def sum(n):
    s=0
    for i in range(1,n+1):
        s=s+i
    return s
n=int(input("Enter u r Number:"))
print(f"original value={n} & Sum is:{sum(n)}")
```

```python
#factorial Of Number
def fact(n):
    f=1
    for i in range(1,n+1):
        f*=i
    return f
n=int(input("Enter u r Number:"))
print(f"original value={n} & factoiral number is ={fact(n)}")
```

```python
#Digital Extraction
def digits(n):
    while n!=0:
        print(n%10)
        n//=10
n=int(input("Enter u r Number:"))
print(f"original value={n}")
digits(n)
```

```python
#Count the digits
def count(n):
    c=0
    while n!=0:
        c=c+1
        n=n//10
    return c
n=int(input("Enter u r Number:"))
print(f"original value={n} & Count Of number is ={count(n)}")
```

```python
#Reverse Number
def rev(n):
    r=0
    while n!=0:
        d=n%10
        r=r*10+d
        n=n//10
    return r
n=int(input("Enter u r Number:"))
print(f"original value={n} & Reverse Of number is ={rev(n)}")
```

```python
#Method 2
def rev(n):
    return str(n)[::-1]
n=int(input("Enter u r Number:"))
print(f"original value={n} & Reverse Of number is ={rev(n)}")
```

```python
#Palindrome
def pal(n):
    t=n
    r=0
    while n!=0:
        d=n%10
        r=r*10+d
        n=n//10
    return t==r
n=int(input("Enter u r Number:"))
print(f"original value={n} & Reverse Of number is ={pal(n)}")
```
